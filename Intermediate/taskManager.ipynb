```python
class Tarefa:
    def __init__(self, descricao):
        self.descricao = descricao
        self.completa = False

    def marcar_como_completa(self):
        self.completa = True
        print(f"A tarefa {self.descricao} foi marcada como completa!")
```

```python
# https://dailypythonprojects.substack.com/p/task-manager-using-object-oriented

class ListaDeTarefas:
    def __init__(self):
        self.tarefas = []

    def adicionar_tarefa(self):
        descricao = input("Qual a descricao da task: ")
        self.tarefas.append(Tarefa(descricao))
        print(f"\nA task {descricao} foi adicionada a lista de a fazeres.\n")

    def completar_tarefa(self):
        self.ver_tarefas()
        num = input("Digite o numero da task que voce deseja completar: ")
        for tarefa in self.tarefas:
            if int(num) == self.tarefas.index(tarefa)+1:
                tarefa.marcar_como_completa()

    def ver_tarefas(self):
        if not self.tarefas:
            print("\nA lista de tasks esta vazia!\n")
        else:
            print("----------------------------")
            print("Imprimindo a lista de tasks:")
            for tarefa in self.tarefas:
                if not tarefa.completa:
                    print(f"{self.tarefas.index(tarefa) + 1}. {tarefa.descricao} : [❌]")
                else:
                    print(f"{self.tarefas.index(tarefa) + 1}. {tarefa.descricao} : [✅]")
            print("----------------------------")

    def run(self):
        while True:
            print("1. Adicionar uma task")
            print("2. Completar task")
            print("3. Ver tasks")
            print("4. Sair")

            escolha = input("Escolha: ")

            if escolha == "1":
                self.adicionar_tarefa()
            elif escolha == "2":
                self.completar_tarefa()
            elif escolha == "3":
                self.ver_tarefas()
            elif escolha == "4":
                print("Voce finalizou o programa!")
                break
            else:
                print("Voce digitou algo errado!")


if __name__ == "__main__":
    listaDeTarefas = ListaDeTarefas()
    listaDeTarefas.run()
```
