```python
class Conta():
    def __init__(self,nome,saldo):
        self.__nome = nome
        self.__saldo = saldo

    def depositar(self,valor):
        self.__saldo += valor

    def verExtrato(self):
        return self.__saldo

    def get_nome(self):
        return self.__nome

    def set_nome(self, nome):
        self.__nome = nome

    def get_saldo(self):
        return self.__saldo

    def set_saldo(self, saldo):
        self.__saldo = saldo
```

```python
# https://dailypythonprojects.substack.com/p/create-a-rectangle-class-oop

class Sistema:

    def __init__(self):
        self.__contas = []

    def menu(self):
        while True:

            print("1.Criar Conta")
            print("2.Depositar dinheiro")
            print("3.Ver extrato")
            print("4.Sair")
            escolha = input("Escolha: ")

            if escolha == "1":
                self.criar_conta()
            elif escolha == "2":
                self.deposito()
            elif escolha == "3":
                self.extrato()
            elif escolha == "4":
                print("Voce finalizou o programa!")
                break
            else:
                print("Voce digitou algo errado!")

    def criar_conta(self):
        nome = input("Digite o seu nome: ")
        saldo = float(input("Digite seu deposito inicial: "))

        conta = Conta(nome, saldo)
        self.__contas.append(conta)

        print(f"Conta criada para {conta.get_nome()} com o saldo inicial de {conta.get_saldo()}", end="\n\n")

    def deposito(self):
        nome = input("Entre com o seu nome: ")
        valor_deposito = float(input("Digite o valor que voce deseja depositar: "))

        if self.__contas == []:
            print("Nao existem contas criadas no banco, crie uma conta primeiro.", end="\n\n")
            return

        for conta in self.__contas:
            if conta.get_nome() == nome:
                conta.depositar(valor_deposito)
                print(f"Voce depositou {valor_deposito} na sua conta.", end="\n\n")

    def extrato(self):
        nome = input("Entre com o seu nome: ")

        if self.__contas == []:
            print("Nao existem contas criadas no banco, crie uma conta primeiro.", end="\n\n")
            return

        for conta in self.__contas:
            if conta.get_nome() == nome:
                print(f"O extrato da conta de {conta.get_nome()} eh {conta.verExtrato()}.", end="\n\n")


if __name__ == '__main__':
    sistema = Sistema()
    sistema.menu()
```
