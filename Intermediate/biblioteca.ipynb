```python
class Livro:
    def __init__(self, titulo):
        self.__titulo = titulo
        self.__disponivel = True

    def marcar_como_alugado(self):
        if self.__disponivel:
            self.__disponivel = False
            print(f"Voce alugou o livro {self.__titulo}")
        else:
            print(f"Desculpe, '{self.__titulo}' ja nao esta disponivel.")

    def marcar_como_retornado(self):
        self.__disponivel = True
        print(f"Obrigado por retornar o livro {self.__titulo}")

    def get_titulo(self):
        return self.__titulo

    def get_disponivel(self):
        return self.__disponivel
```

```python
# https://dailypythonprojects.substack.com/p/copy-banking-system-with-python-using

class Biblioteca:
    def __init__(self):
        self.livros = []

    def add_book(self):
        titulo = input("Digite o titulo do livro que voce deseja adicionar a biblioteca: ")
        self.livros.append(Livro(titulo))
        print(f"Voce adicionou o livro {titulo} com sucesso!")

    def alugar_livro(self):
        titulo = input("Digite o titulo do livro que voce deseja alugar: ")
        for livro in self.livros:
            if livro.get_titulo() == titulo:
                    livro.marcar_como_alugado()

    def retornar_livro(self):
        titulo = input("Digite o titulo do livro que voce deseja retornar: ")
        for livro in self.livros:
            if livro.get_titulo() == titulo:
                livro.marcar_como_retornado()

    def ver_livros(self):
        if not self.livros:
            print("Nao ha livros disponiveis para aluguel!")
        else:
            i = 1
            print("Os livros disponiveis para aluguel s達o:")
            for livro in self.livros:
                if livro.get_disponivel():
                    print(f"{i}. {livro.get_titulo()}")
                    i += 1


    def run(self):
        while True:
            print("\n1. Adicionar Livro\n2. Alugar Livro\n3. Retornar Livro\n4. Ver Livros Disponiveis\n5. Sair")
            escolha = input("Digite sua escolha: ")

            if escolha == '1':
                self.add_book()
            elif escolha == '2':
                self.alugar_livro()
            elif escolha == '3':
                self.retornar_livro()
            elif escolha == '4':
                self.ver_livros()
            elif escolha == '5':
                print("Voce saiu da biblioteca.")
                break
            else:
                print("Voce digitou uma opcao invalida!")

if __name__ == "__main__":
    biblio = Biblioteca()
    biblio.run()
```
